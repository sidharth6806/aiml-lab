```python
import numpy as np
import pandas as pd
data = {
    "type_of_house":np.array(['pg','flat','room','pg','flat','room','pg','flat','room','pg','flat','room','pg','flat','room','pg','flat','room',]),
    "area_of_house":np.array([24,48,18,26,50,19,29,56,21,29,53,13,11,34,23,23,11,55]),
    "rooms_size":np.array([2,4,3,2,5,2,1,2,1,1,3,2,1,3,4,2,3,4,]),
    "cost":np.array([100000,300000,200000,50000,100000,25000,150000,275000,175000,100000,300000,200000,50000,100000,25000,150000,275000,175000])
}
df = pd.DataFrame(data)
print(df)



```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x="area_of_house",y= "cost",hue="type_of_house",data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x="type_of_house",y="cost",hue="area_of_house",data=df)
plt.show()
```

```python
avg_cost = df.groupby("type_of_house")["cost"].mean()

plt.figure()
plt.bar(avg_cost.index, avg_cost.values)
plt.xlabel("Type of House")
plt.ylabel("Average Cost")
plt.title("Average Cost per House Type")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.boxplot(data=df, x="type_of_house", y="cost")
plt.title("Cost Distribution by House Type")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#plt.figure()
sns.scatterplot(x="area_of_house", y="cost", data=df)
plt.title("Area vs Cost")
plt.show()
```

```python
plt.figure()
sns.barplot(x="type_of_house", y="cost", data=df)
plt.title("Average Cost by Type")
plt.show()
```

```python
import numpy as np

# Sort by area
df_sorted = df.sort_values("area_of_house")

x = np.array(df_sorted["area_of_house"])
y = np.array(df_sorted["cost"])

plt.figure()
plt.plot(x, y)
plt.title("Area vs Cost Line Plot")
plt.xlabel("Area")
plt.ylabel("Cost")
plt.show()
```
