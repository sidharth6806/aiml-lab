```python
import numpy as np
import pandas as pd
data = {
    "type_of_house":np.array(['pg','flat','room','pg','flat','room','pg','flat','room','pg','flat','room','pg','flat','room','pg','flat','room',]),
    "area_of_house":np.array([24,48,18,26,50,19,29,56,21,29,53,13,11,34,23,23,11,55]),
    "rooms_size":np.array([2,4,3,2,5,2,1,2,1,1,3,2,1,3,4,2,3,4,]),
    "cost":np.array([100000,300000,200000,50000,100000,25000,150000,275000,175000,100000,300000,200000,50000,100000,25000,150000,275000,175000])
}
df = pd.DataFrame(data)
print(df)
```

```python
listarea= df["area_of_house"].tolist()
minval = min(listarea)
maxval = max(listarea)

print("Minimum:", minval)
print("Maximum:", maxval)

area = []

for x in listarea:
    value = (x - minval) / (maxval - minval)
    area.append(value)

for value in area:
    print(round(value, 3))

```

```python
room_list = df["rooms_size"].tolist()
min_room = min(room_list)
max_room = max(room_list)

print("Min:", min_room)
print("Max:", max_room)
print() 
room = []

for x in room_list:
    value = (x - min_room) / (max_room - min_room)
    room.append(round(value, 3))

for value in room:
    print(value)
```

```python
costt = df["cost"].tolist()

min_cost = min(costt)
max_cost = max(costt)

print("Min:", min_cost)
print("Max:", max_cost)
print()

cost= []

for x in costt:
    value = (x - min_cost) / (max_cost - min_cost)
    cost.append(round(value, 3))

for value in cost:
    print(value)
```
